```python
from google.colab import drive
drive.mount('/content/drive')

import json
import random
from pathlib import Path

import numpy as np
import pandas as pd
import plotly.graph_objects as go
import tensorflow as tf
from plotly.subplots import make_subplots
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
import matplotlib.pyplot as plt
pd.set_option('display.max_colwidth', 120)

# =========================
# Config
# =========================
DATA_NPZ_PATH = Path('/content/drive/MyDrive/predprof/Data.npz')
OUTPUT_DIR = Path('/content/drive/MyDrive/predprof/')
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

SEED = 42
SAMPLE_RATE = 16000
EXPECTED_NUM_SAMPLES = 80000

FRAME_LENGTH = 1024
FRAME_STEP = 256
FFT_LENGTH = 1024
N_MELS = 96
IMG_HEIGHT = 96
IMG_WIDTH = 160

BATCH_SIZE = 32
EPOCHS = 40
LEARNING_RATE = 3e-4

USE_WAVE_AUG = True
USE_SPEC_AUG = True
WAVE_SHIFT_PCT = 0.10
GAIN_LOW = 0.85
GAIN_HIGH = 1.15
NOISE_STD = 0.002

assert DATA_NPZ_PATH.exists(), f'File not found: {DATA_NPZ_PATH}'

random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

print('TensorFlow version:', tf.__version__)
print('GPU devices:', tf.config.list_physical_devices('GPU'))
print('DATA_NPZ_PATH =', DATA_NPZ_PATH)
print('OUTPUT_DIR =', OUTPUT_DIR)

# =========================
# Load data
# =========================
npz = np.load(DATA_NPZ_PATH, allow_pickle=False)
print('npz keys:', npz.files)

train_x = npz['train_x'].astype(np.float32).squeeze(-1)
valid_x = npz['valid_x'].astype(np.float32).squeeze(-1)
train_raw_y = npz['train_y'].astype(str)
valid_raw_y = npz['valid_y'].astype(str)

assert train_x.ndim == 2 and valid_x.ndim == 2
assert train_x.shape[1] == EXPECTED_NUM_SAMPLES, train_x.shape
assert valid_x.shape[1] == EXPECTED_NUM_SAMPLES, valid_x.shape

def split_raw_label(raw_label: str):
    raw_label = str(raw_label)
    uid32 = raw_label[:32]
    label = raw_label[32:]
    return uid32, label

def build_meta(raw_labels, split_name):
    rows = []
    for i, raw_label in enumerate(raw_labels):
        uid32, label = split_raw_label(raw_label)
        rows.append({
            'split': split_name,
            'index': i,
            'raw_label': raw_label,
            'uid32': uid32,
            'label': label,
        })
    return pd.DataFrame(rows)

train_meta = build_meta(train_raw_y, 'train')
valid_meta = build_meta(valid_raw_y, 'valid')
meta = pd.concat([train_meta, valid_meta], ignore_index=True)

class_names = sorted(meta['label'].astype(str).unique().tolist())
label_to_id = {label: idx for idx, label in enumerate(class_names)}
id_to_label = {idx: label for label, idx in label_to_id.items()}
NUM_CLASSES = len(class_names)

train_meta['class_id'] = train_meta['label'].map(label_to_id)
valid_meta['class_id'] = valid_meta['label'].map(label_to_id)
y_train = train_meta['class_id'].to_numpy(np.int32)
y_val = valid_meta['class_id'].to_numpy(np.int32)

assert (train_meta['raw_label'].str[:32] == train_meta['uid32']).all()
assert (train_meta['raw_label'].str[32:] == train_meta['label']).all()
assert (valid_meta['raw_label'].str[:32] == valid_meta['uid32']).all()
assert (valid_meta['raw_label'].str[32:] == valid_meta['label']).all()

uid_overlap = len(set(train_meta['uid32']) & set(valid_meta['uid32']))
print('train_x shape:', train_x.shape)
print('valid_x shape:', valid_x.shape)
print('num_classes:', NUM_CLASSES)
print('chance accuracy:', round(1.0 / NUM_CLASSES, 4))
print('uid overlap train/valid:', uid_overlap)

pd.DataFrame({'class_id': range(NUM_CLASSES), 'label': class_names}).to_csv(OUTPUT_DIR / 'tiny_cnn_label_mapping.csv', index=False)

# =========================
# tf.data pipeline
# =========================
AUTOTUNE = tf.data.AUTOTUNE
NUM_SPECTROGRAM_BINS = FFT_LENGTH // 2 + 1
MEL_WEIGHT_MATRIX = tf.signal.linear_to_mel_weight_matrix(
    num_mel_bins=N_MELS,
    num_spectrogram_bins=NUM_SPECTROGRAM_BINS,
    sample_rate=SAMPLE_RATE,
    lower_edge_hertz=20.0,
    upper_edge_hertz=SAMPLE_RATE / 2.0,
)

def normalize_waveform(wave):
    wave = tf.cast(wave, tf.float32)
    wave = tf.reshape(wave, [EXPECTED_NUM_SAMPLES])
    max_abs = tf.reduce_max(tf.abs(wave))
    return tf.cond(max_abs > 0, lambda: wave / max_abs, lambda: wave)

def augment_waveform(wave):
    max_shift = int(WAVE_SHIFT_PCT * EXPECTED_NUM_SAMPLES)
    if max_shift > 0:
        shift = tf.random.uniform([], minval=-max_shift, maxval=max_shift + 1, dtype=tf.int32)
        wave = tf.roll(wave, shift=shift, axis=0)
    gain = tf.random.uniform([], GAIN_LOW, GAIN_HIGH)
    wave = wave * gain
    wave = wave + tf.random.normal(tf.shape(wave), stddev=NOISE_STD)
    return tf.clip_by_value(wave, -1.0, 1.0)

def waveform_to_logmel(wave):
    stft = tf.signal.stft(
        wave,
        frame_length=FRAME_LENGTH,
        frame_step=FRAME_STEP,
        fft_length=FFT_LENGTH,
        pad_end=True,
    )
    power = tf.abs(stft) ** 2
    mel = tf.matmul(power, MEL_WEIGHT_MATRIX)
    logmel = tf.math.log(mel + 1e-6)
    logmel = tf.transpose(logmel, perm=[1, 0])
    mean = tf.reduce_mean(logmel)
    std = tf.math.reduce_std(logmel)
    logmel = (logmel - mean) / (std + 1e-6)
    image = tf.expand_dims(logmel, axis=-1)
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    return image

def specaugment(image):
    h = tf.shape(image)[0]
    w = tf.shape(image)[1]
    t = tf.random.uniform([], 0, tf.maximum(2, w // 8), dtype=tf.int32)
    ts = tf.random.uniform([], 0, tf.maximum(1, w - t + 1), dtype=tf.int32)
    time_mask = tf.concat([
        tf.ones([h, ts, 1], dtype=image.dtype),
        tf.zeros([h, t, 1], dtype=image.dtype),
        tf.ones([h, w - ts - t, 1], dtype=image.dtype),
    ], axis=1)
    image = image * time_mask
    f = tf.random.uniform([], 0, tf.maximum(2, h // 8), dtype=tf.int32)
    fs = tf.random.uniform([], 0, tf.maximum(1, h - f + 1), dtype=tf.int32)
    freq_mask = tf.concat([
        tf.ones([fs, w, 1], dtype=image.dtype),
        tf.zeros([f, w, 1], dtype=image.dtype),
        tf.ones([h - fs - f, w, 1], dtype=image.dtype),
    ], axis=0)
    return image * freq_mask

def preprocess_train(wave, label):
    wave = normalize_waveform(wave)
    if USE_WAVE_AUG:
        wave = augment_waveform(wave)
    image = waveform_to_logmel(wave)
    if USE_SPEC_AUG:
        image = specaugment(image)
    return image, label

def preprocess_eval(wave, label):
    wave = normalize_waveform(wave)
    image = waveform_to_logmel(wave)
    return image, label

def build_dataset(waves, labels, training=False):
    ds = tf.data.Dataset.from_tensor_slices((waves, labels))
    if training:
        ds = ds.shuffle(len(waves), seed=SEED, reshuffle_each_iteration=True)
        ds = ds.map(preprocess_train, num_parallel_calls=AUTOTUNE)
    else:
        ds = ds.map(preprocess_eval, num_parallel_calls=AUTOTUNE)
        ds = ds.cache()
    return ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)

train_ds = build_dataset(train_x, y_train, training=True)
val_ds = build_dataset(valid_x, y_val, training=False)
sample_batch, sample_labels = next(iter(val_ds.take(1)))
print('sample batch shape:', sample_batch.shape)

fig, axes = plt.subplots(2, 2, figsize=(12, 6))
for i in range(2):
    axes[i, 0].plot(train_x[i])
    axes[i, 0].set_title(f'Waveform | {id_to_label[int(y_train[i])]}')
    axes[i, 1].imshow(sample_batch[i, :, :, 0], aspect='auto', origin='lower')
    axes[i, 1].set_title(f'Log-mel | {id_to_label[int(sample_labels[i])]}')
plt.tight_layout()
plt.show()

# =========================
# Small CNN
# =========================
layers = tf.keras.layers

def conv_block(x, filters, dropout_rate):
    x = layers.Conv2D(filters, 3, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(filters, 3, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)
    x = layers.Dropout(dropout_rate)(x)
    return x

inputs = layers.Input(shape=sample_batch.shape[1:])
x = conv_block(inputs, 32, 0.10)
x = conv_block(x, 64, 0.12)
x = conv_block(x, 128, 0.15)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.15)(x)
outputs = layers.Dense(NUM_CLASSES, activation='softmax', dtype='float32')(x)
model = tf.keras.Model(inputs, outputs, name='tiny_audio_cnn')

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy(name='accuracy')],
)
model.summary()

BEST_MODEL_PATH = OUTPUT_DIR / 'tiny_cnn_best_model.keras'
H5_MODEL_PATH = OUTPUT_DIR / 'tiny_cnn_model.h5'
HISTORY_CSV_PATH = OUTPUT_DIR / 'tiny_cnn_history.csv'
PRED_CSV_PATH = OUTPUT_DIR / 'tiny_cnn_validation_predictions.csv'
CONFIG_PATH = OUTPUT_DIR / 'tiny_cnn_config.json'
ANALYTICS_JSON_PATH = OUTPUT_DIR / 'tiny_cnn_analytics.json'
TRAINING_CURVES_HTML = OUTPUT_DIR / 'tiny_cnn_training_curves.html'
CLASS_DISTRIBUTION_HTML = OUTPUT_DIR / 'tiny_cnn_class_distribution.html'
VALIDATION_CONFIDENCE_HTML = OUTPUT_DIR / 'tiny_cnn_validation_confidence.html'
TOP5_VALID_HTML = OUTPUT_DIR / 'tiny_cnn_top5_validation_classes.html'
CONFUSION_MATRIX_HTML = OUTPUT_DIR / 'tiny_cnn_confusion_matrix.html'

callbacks = [
    tf.keras.callbacks.ModelCheckpoint(str(BEST_MODEL_PATH), monitor='val_accuracy', mode='max', save_best_only=True, verbose=1),
    tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', mode='max', patience=8, restore_best_weights=True, verbose=1),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1),
]

history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, callbacks=callbacks, verbose=1)
history_df = pd.DataFrame(history.history)
history_df.to_csv(HISTORY_CSV_PATH, index=False)

best_model = tf.keras.models.load_model(str(BEST_MODEL_PATH))
best_model.save(str(H5_MODEL_PATH), include_optimizer=False)

train_eval_ds = build_dataset(train_x, y_train, training=False)
train_probs = best_model.predict(train_eval_ds, verbose=0)
val_probs = best_model.predict(val_ds, verbose=0)
train_pred = train_probs.argmax(axis=1)
val_pred = val_probs.argmax(axis=1)

train_loss_eval, train_acc_eval = best_model.evaluate(train_eval_ds, verbose=0)
val_loss_eval, val_acc_eval = best_model.evaluate(val_ds, verbose=0)

train_acc = accuracy_score(y_train, train_pred)
val_acc = accuracy_score(y_val, val_pred)
train_macro_f1 = f1_score(y_train, train_pred, average='macro')
val_macro_f1 = f1_score(y_val, val_pred, average='macro')

print(f'Train accuracy: {train_acc:.4f}')
print(f'Train macro-F1: {train_macro_f1:.4f}')
print(f'Val accuracy:   {val_acc:.4f}')
print(f'Val macro-F1:   {val_macro_f1:.4f}')

print('\nValidation classification report:\n')
print(classification_report(y_val, val_pred, labels=np.arange(NUM_CLASSES), target_names=class_names, zero_division=0))

pred_diag_df = pd.DataFrame({
    'class_id': np.arange(NUM_CLASSES),
    'label': class_names,
    'val_true_count': np.bincount(y_val, minlength=NUM_CLASSES),
    'val_pred_count': np.bincount(val_pred, minlength=NUM_CLASSES),
    'mean_val_prob': val_probs.mean(axis=0),
})
display(pred_diag_df.sort_values('val_pred_count', ascending=False))

val_predictions_df = valid_meta.copy()
val_predictions_df['true_class_id'] = y_val
val_predictions_df['pred_class_id'] = val_pred
val_predictions_df['true_label'] = val_predictions_df['true_class_id'].map(id_to_label)
val_predictions_df['pred_label'] = [id_to_label[int(x)] for x in val_pred]
val_predictions_df['confidence'] = val_probs.max(axis=1)
val_predictions_df['correct'] = val_predictions_df['true_class_id'] == val_predictions_df['pred_class_id']
val_predictions_df.to_csv(PRED_CSV_PATH, index=False)

theme_color = '#00BCD4'
grid_color = '#1DAAE6'
bg_color = '#000066'
title_font_family = 'Arial Black'
body_font_family = 'Roboto, sans-serif'

def style_figure(fig, title, xaxis_title, yaxis_title, width=1000, height=600, hovermode='x'):
    fig.update_layout(
        title={
            'text': title,
            'y': 0.95,
            'x': 0.5,
            'xanchor': 'center',
            'yanchor': 'top',
            'font': dict(size=24, family=title_font_family, color=theme_color),
        },
        xaxis_title=xaxis_title,
        yaxis_title=yaxis_title,
        paper_bgcolor=bg_color,
        plot_bgcolor=bg_color,
        hovermode=hovermode,
        bargap=0.1,
        height=height,
        width=width,
        font=dict(family=body_font_family, size=14, color=theme_color),
        legend=dict(font=dict(color=theme_color)),
    )
    fig.update_xaxes(showgrid=False, gridcolor=grid_color, linecolor=theme_color, tickfont=dict(size=10, color=theme_color), title_font=dict(color=theme_color))
    fig.update_yaxes(showgrid=True, gridcolor=grid_color, linecolor=theme_color, tickfont=dict(color=theme_color), title_font=dict(color=theme_color))
    return fig

def show_and_save(fig, path):
    fig.show()
    fig.write_html(str(path), include_plotlyjs='cdn')

training_fig = make_subplots(
    rows=1,
    cols=2,
    subplot_titles=('Train vs Val Loss', 'Train vs Val Accuracy'),
    horizontal_spacing=0.12,
)
training_fig.add_trace(go.Scatter(x=np.arange(1, len(history_df) + 1), y=history_df['loss'], mode='lines+markers', name='Train Loss', line=dict(color='#00E5FF', width=3)), row=1, col=1)
training_fig.add_trace(go.Scatter(x=np.arange(1, len(history_df) + 1), y=history_df['val_loss'], mode='lines+markers', name='Val Loss', line=dict(color='#80DEEA', width=3)), row=1, col=1)
training_fig.add_trace(go.Scatter(x=np.arange(1, len(history_df) + 1), y=history_df['accuracy'], mode='lines+markers', name='Train Accuracy', line=dict(color='#26C6DA', width=3)), row=1, col=2)
training_fig.add_trace(go.Scatter(x=np.arange(1, len(history_df) + 1), y=history_df['val_accuracy'], mode='lines+markers', name='Val Accuracy', line=dict(color='#B2EBF2', width=3)), row=1, col=2)
style_figure(training_fig, 'Кривые обучения модели', 'Эпоха', 'Значение', width=1200, height=600)
training_fig.update_annotations(font=dict(size=16, family=body_font_family, color=theme_color))
training_fig.update_xaxes(tickmode='linear', dtick=1)
show_and_save(training_fig, TRAINING_CURVES_HTML)

train_distribution_df = train_meta[['class_id']].copy()
class_dist_fig = go.Figure()
class_dist_fig.add_trace(go.Histogram(
    x=train_distribution_df['class_id'],
    name='class_id',
    marker_color=theme_color,
    opacity=0.75,
    hovertemplate='<b>Категория:</b> %{x}<br><b>Частота:</b> %{y}<extra></extra>',
))
style_figure(class_dist_fig, 'Распределение классов в train', 'ID Категории', 'Количество записей')
class_dist_fig.update_xaxes(
    tickmode='linear',
    dtick=1,
    range=[int(train_distribution_df['class_id'].min()) - 0.5, int(train_distribution_df['class_id'].max()) + 0.5],
    tickangle=45,
)
show_and_save(class_dist_fig, CLASS_DISTRIBUTION_HTML)

val_predictions_df['record_id'] = np.arange(1, len(val_predictions_df) + 1)
confidence_fig = go.Figure()
confidence_fig.add_trace(go.Scatter(
    x=val_predictions_df['record_id'],
    y=val_predictions_df['confidence'],
    mode='markers',
    name='Confidence',
    marker=dict(color=theme_color, size=8, opacity=0.75),
    customdata=val_predictions_df[['true_class_id', 'pred_class_id', 'correct']].to_numpy(),
    hovertemplate='<b>Запись:</b> %{x}<br><b>Confidence:</b> %{y:.4f}<br><b>True:</b> %{customdata[0]}<br><b>Pred:</b> %{customdata[1]}<br><b>Correct:</b> %{customdata[2]}<extra></extra>',
))
style_figure(confidence_fig, 'Уверенность модели по каждой validation-записи', 'Номер записи', 'Confidence')
confidence_fig.update_yaxes(range=[0, 1])
show_and_save(confidence_fig, VALIDATION_CONFIDENCE_HTML)

top5_valid_df = (
    valid_meta['class_id']
    .value_counts()
    .sort_values(ascending=False)
    .head(5)
    .rename_axis('class_id')
    .reset_index(name='count')
)
top5_valid_df['label'] = top5_valid_df['class_id'].map(id_to_label)
top5_fig = go.Figure()
top5_fig.add_trace(go.Bar(
    x=top5_valid_df['class_id'],
    y=top5_valid_df['count'],
    marker_color=theme_color,
    opacity=0.8,
    customdata=top5_valid_df[['label']].to_numpy(),
    hovertemplate='<b>ID:</b> %{x}<br><b>Label:</b> %{customdata[0]}<br><b>Количество:</b> %{y}<extra></extra>',
))
style_figure(top5_fig, 'Топ-5 самых частых классов в validation', 'ID Категории', 'Количество записей')
top5_fig.update_xaxes(tickmode='linear', dtick=1, tickangle=45)
show_and_save(top5_fig, TOP5_VALID_HTML)

cm = confusion_matrix(y_val, val_pred, labels=np.arange(NUM_CLASSES))
cm_fig = go.Figure(data=go.Heatmap(
    z=cm,
    x=np.arange(NUM_CLASSES),
    y=np.arange(NUM_CLASSES),
    colorscale='Blues',
    hovertemplate='<b>Pred:</b> %{x}<br><b>True:</b> %{y}<br><b>Count:</b> %{z}<extra></extra>',
    colorbar=dict(title='Count'),
))
style_figure(cm_fig, 'Validation confusion matrix', 'Predicted class_id', 'True class_id', hovermode='closest')
show_and_save(cm_fig, CONFUSION_MATRIX_HTML)

analytics_payload = {
    'epochs': np.arange(1, len(history_df) + 1).tolist(),
    'train_loss_curve': history_df['loss'].astype(float).tolist(),
    'val_loss_curve': history_df['val_loss'].astype(float).tolist(),
    'train_accuracy_curve': history_df['accuracy'].astype(float).tolist(),
    'val_accuracy_curve': history_df['val_accuracy'].astype(float).tolist(),
    'train_class_counts': train_meta['class_id'].value_counts().sort_index().astype(int).to_dict(),
    'validation_confidence': val_predictions_df[['record_id', 'true_class_id', 'pred_class_id', 'confidence', 'correct']].to_dict(orient='records'),
    'top5_validation_classes': top5_valid_df.to_dict(orient='records'),
    'final_accuracy': float(val_acc_eval),
    'final_loss': float(val_loss_eval),
    'class_names': class_names,
}
with ANALYTICS_JSON_PATH.open('w', encoding='utf-8') as f:
    json.dump(analytics_payload, f, ensure_ascii=False, indent=2)

config = {
    'data_npz_path': str(DATA_NPZ_PATH),
    'sample_rate': SAMPLE_RATE,
    'expected_num_samples': EXPECTED_NUM_SAMPLES,
    'frame_length': FRAME_LENGTH,
    'frame_step': FRAME_STEP,
    'fft_length': FFT_LENGTH,
    'n_mels': N_MELS,
    'img_height': IMG_HEIGHT,
    'img_width': IMG_WIDTH,
    'batch_size': BATCH_SIZE,
    'epochs': EPOCHS,
    'learning_rate': LEARNING_RATE,
    'use_wave_aug': USE_WAVE_AUG,
    'use_spec_aug': USE_SPEC_AUG,
    'num_classes': NUM_CLASSES,
    'class_names': class_names,
    'train_accuracy': float(train_acc),
    'val_accuracy': float(val_acc),
    'train_macro_f1': float(train_macro_f1),
    'val_macro_f1': float(val_macro_f1),
    'train_loss_eval': float(train_loss_eval),
    'val_loss_eval': float(val_loss_eval),
}
with CONFIG_PATH.open('w', encoding='utf-8') as f:
    json.dump(config, f, ensure_ascii=False, indent=2)

print('\nSaved artifacts:')
for name in [
    'tiny_cnn_label_mapping.csv',
    'tiny_cnn_best_model.keras',
    'tiny_cnn_model.h5',
    'tiny_cnn_history.csv',
    'tiny_cnn_validation_predictions.csv',
    'tiny_cnn_config.json',
    'tiny_cnn_analytics.json',
    'tiny_cnn_training_curves.html',
    'tiny_cnn_class_distribution.html',
    'tiny_cnn_validation_confidence.html',
    'tiny_cnn_top5_validation_classes.html',
    'tiny_cnn_confusion_matrix.html',
]:
    print(' -', OUTPUT_DIR / name)
```
